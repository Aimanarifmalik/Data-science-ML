```python
import pandas as pd
import matplotlib.pyplot as plt

#1 load dataset
df = pd.read_excel('WorldBank.xlsx')
df.head()


#2histograms to visualize the distribution of a continuous variable.
# Loop through each column and create histograms for continuous numeric values
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        plt.figure(figsize=(10, 6))
        plt.hist(df[column].dropna(), bins=30, edgecolor='black')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.xticks(rotation=45)  # Rotate x-axis labels if needed
        plt.show()





```

```python
import pandas as pd
from scipy.stats import skew, kurtosis


df = pd.read_excel('WorldBank.xlsx')

# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Calculate skewness and kurtosis for each numeric column
skewness = numeric_df.apply(skew)
kurtosis_values = numeric_df.apply(kurtosis)


print("Skewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis_values)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_excel('WorldBank.xlsx')
print(df.head())


print("Column names in the DataFrame:")
print(df.columns)

# Assuming 'Education Expenditure (% of GDP)' and 'GDP Growth Rate' are columns in your dataset
# Adjust these column names based on the actual column names in your dataset
education_expenditure_col = 'Education Expenditure (% of GDP)'  # Replace with actual column name if different
gdp_growth_rate_col = 'GDP Growth Rate'  # Replace with actual column name if different

if education_expenditure_col in df.columns and gdp_growth_rate_col in df.columns:
    education_expenditure = df[education_expenditure_col]
    gdp_growth_rate = df[gdp_growth_rate_col]

    # Define threshold for high and low education expenditure
    threshold = education_expenditure.median()

    # Create two groups based on the threshold
    high_education_expenditure = gdp_growth_rate[education_expenditure > threshold]
    low_education_expenditure = gdp_growth_rate[education_expenditure <= threshold]

    # Drop NaN values
    high_education_expenditure = high_education_expenditure.dropna()
    low_education_expenditure = low_education_expenditure.dropna()

    # Perform t-test for independent samples
    t_statistic, p_value = ttest_ind(high_education_expenditure, low_education_expenditure)

    # Print t-statistic and p-value
    print(f"T-statistic: {t_statistic}")
    print(f"P-value: {p_value}")

    # Interpret the results based on the p-value (typically, if p-value < 0.05, reject H₀)
    if p_value < 0.05:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")
else:
    print(f"Columns {education_expenditure_col} and/or {gdp_growth_rate_col} not found in the DataFrame.")
```

```python
import pandas as pd
from scipy.stats import pearsonr


df = pd.read_excel('WorldBank.xlsx')
print(df.head())


print("Column names in the DataFrame:")
print(df.columns)

# Assuming 'Education Expenditure (% of GDP)' and 'GDP Growth Rate' are columns in your dataset
# Adjust these column names based on the actual column names in your dataset
education_expenditure_col = 'Education Expenditure (% of GDP)'  # Replace with actual column name if different
gdp_growth_rate_col = 'GDP Growth Rate'  # Replace with actual column name if different

if education_expenditure_col in df.columns and gdp_growth_rate_col in df.columns:
    education_expenditure = df[education_expenditure_col]
    gdp_growth_rate = df[gdp_growth_rate_col]

    # Drop rows with NaN values from both series
    valid_data = df[[education_expenditure_col, gdp_growth_rate_col]].dropna()

    education_expenditure = valid_data[education_expenditure_col]
    gdp_growth_rate = valid_data[gdp_growth_rate_col]

    # Perform Pearson correlation coefficient and p-value calculation
    corr, p_value = pearsonr(education_expenditure, gdp_growth_rate)

   
    print(f"Pearson correlation coefficient: {corr}")
    print(f"P-value: {p_value}")

    # Interpret the results based on the p-value (typically, if p-value < 0.05, reject H₀)
    if p_value < 0.05:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")
else:
    print(f"Columns {education_expenditure_col} and/or {gdp_growth_rate_col} not found in the DataFrame.")
```
