```python
import pandas as pd
df=pd.read_csv('train.csv')
df.head(5)
```

2. Visualization of Numerical Features:

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['SibSp'].astype(str)
df['Parch'].astype(str)
df['FamilySize']=df['SibSp'].astype(str).str.cat(df['Parch'].astype(str), sep="_")
print(df[['FamilySize']])

sns.histplot(df['FamilySize'].dropna(),kde=True)
plt.title('FamilySize Distribution')
plt.show()
sns.histplot(df['Age'].dropna(),kde=True)
plt.title('Age Distribution')
plt.show()
sns.histplot(df['Fare'].dropna(),kde=True)
plt.title('Fare Distribution')
plt.show()
```

3. Feature Relationships:

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df['Age'],y=df['Fare'])
plt.title('Fare and Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#selecting only numeric columns
numeric_col=df.select_dtypes(include=['number'])
corr_mat=numeric_col.corr()
sns.heatmap(corr_mat,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

 Exploratory Data Analysis II

```python
df['Title']=df['Name'].str.extract('([A-Za-z]+)\.',expand=False)
plt.figure(figsize=(10, 6))
sns.countplot(x='Title',data=df)
plt.title('Title Distribution')
plt.xlabel('Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='Pclass',data=df)
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='Embarked',data=df)
plt.title('Plot of Embarkation Distribution')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Pclass',y='Fare',data=df)
plt.title('Violin Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Survived',y='Age',data=df)
plt.title('Swarm Plot of Age by Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

plt.figure(figsize=(10, 6))
sns.swarmplot(x='Survived',y='Fare',data=df)
plt.title('Swarm Plot of Fare by Survived')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

```python
features=['Name','Age','Fare','SibSp','Sex','Embarked']
sns.pairplot(df[features])
plt.show()
```
