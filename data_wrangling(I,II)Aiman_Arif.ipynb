Task 1: Data Wrangling I

```python
import pandas as pd
df=pd.read_csv('test.csv')
df.dtypes
df.shape
df.head(10)
```

```python
#remove leading and trailing characters
df.columns=df.columns.str.strip()
df
```

Handle Missing Values:

```python
df.isnull()
mean=df['Age'].mean()
print(df['Age'].fillna(mean,inplace=True))

mode=df['Embarked'].mode()
print(df['Embarked'].fillna(mode,inplace=True))

print(df.dropna(subset=['Cabin'], inplace=True))
print(df)



```

3. Data Type Conversion:
   - Convert the 'Pclass' column to a categorical data type

```python
df['Pclass'] = df['Pclass'].astype('category')
print(df)
```

4. Normalization/Scaling:
   - Normalize the 'Fare' column using Min-Max scaling.

```python
min=df['Fare'].min()
max=df['Fare'].max()
df['Fare']=(df['Fare']-min)/(max-min)
print("\n\nDataFrame after Min-Max scaling the fare column:",df)
```

5. Encoding Categorical Variables:
   - One-hot encode the 'Sex' and 'Embarked' columns

```python
encoding=pd.get_dummies(df,columns=['Sex','Embarked'])

print("\nDataFrame after one-hot encoding sex and Embarked column:",encoding)
```

Task 2: Data Wrangling II

```python
import pandas as pd
df=pd.read_csv('train.csv')
df.head(5)
```

1. Feature Creation:

```python
df['SibSp'].astype(str)
df['Parch'].astype(str)
df['FamilySize']=df['SibSp'].astype(str).str.cat(df['Parch'].astype(str), sep="_")
```

```python
df['IsAlone']=(df['FamilySize']==1).astype(int)

print("DataFrame with IsAlone column:")
print(df)
```

2. Log Transformation:
   - Apply a log transformation to the 'Fare' column to handle skewness.

```python
import numpy as np
df['Fare_log']=np.log1p(df['Fare'])
print(df)
```

3. Regular Expressions:
   - Extract titles from the 'Name' column and create a new column 'Title'.

```python
# ^ shows that the title must be at the beginning of the string
#(Mrs\.|Miss|Dr) matches any of the titles Mr,Mrs,Mis,Dr
df['Title']=df['Name'].str.extract('^(Mr.|Mrs.|Miss|Dr)')
print("DataFrame with title column:",df)
```

4. Duplicate Handling and Outliers:

```python
df = df.drop_duplicates()
numeric_col=['Fare', 'Age']  
numeric_df = df[numeric_col]

#detect and handle outliers using the IQR method
Q1=numeric_df.quantile(0.25)
Q3=numeric_df.quantile(0.75)
IQR=Q3-Q1

#filter out the outliers
outlier_iqr = (numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))
df_cleaned = df[~outlier_iqr.any(axis=1)]
print(df_cleaned)
```
