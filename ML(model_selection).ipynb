1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.isnull()
```

```python
df['Sex'].replace({'male': 0, 'female': 1}, inplace=True)
mode_cabin =df['Cabin'].mode()[0]
df['Cabin'].fillna(mode_cabin, inplace=True)   
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
X = df[['Pclass', 'Sex', 'Age']]
Y = df['Survived']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python

log_reg = LogisticRegression(max_iter=1000)
```

```python
cv_scores_log_reg = cross_val_score(log_reg, X, Y, cv=5)
print(f'Logistic Regression - Cross-Validation Scores: {cv_scores_log_reg}')
print(f'Logistic Regression - Mean CV Score: {cv_scores_log_reg.mean():.2f}')
print(f'Logistic Regression - Standard Deviation of CV Scores: {cv_scores_log_reg.std():.2f}')
```

2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import pandas as pd


df = pd.read_csv('train.csv')
```

```python
df.isnull()
```

```python
df['Sex'].replace({'male': 0, 'female': 1}, inplace=True)
mode_cabin= df['Cabin'].mode()[0]
df['Cabin'].fillna(mode_cabin, inplace=True)   
```

```python
# Features and target variable
X = df[['Pclass', 'Sex', 'Age']]
Y = df['Survived']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
decision_tree = DecisionTreeClassifier()
```

```python
cv_scores_decision_tree = cross_val_score(decision_tree, X, Y, cv=5)
print('Decision Tree - Cross-Validation Scores: {cv_scores_decision_tree}')
print('Decision Tree - Mean CV Score: {cv_scores_decision_tree.mean():}')
print('Decision Tree - Standard Deviation of CV Scores: {cv_scores_decision_tree.std():}')
```

3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import pandas as pd


df = pd.read_csv('train.csv')
```

```python
df.isnull()
```

```python
df['Sex'].replace({'male': 0, 'female': 1}, inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)  # Fill missing Age with median
df['Cabin'].fillna('Unknown', inplace=True)      
```

```python
X = df[['Pclass', 'Sex', 'Age']]
Y = df['Survived']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
random_forest = RandomForestClassifier()
```

```python
cv_scores_random_forest = cross_val_score(random_forest, X, Y, cv=5)


print('Random Forest - Cross-Validation Scores: {cv_scores_random_forest}')
print('Random Forest - Mean CV Score: {cv_scores_random_forest.mean():.2f}')
print('Random Forest - Standard Deviation of CV Scores: {cv_scores_random_forest.std():.2f}')
```

4. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import pandas as pd


df = pd.read_csv('train.csv')
```

```python
df.isnull()
```

```python
# Preprocessing
df['Sex'].replace({'male': 0, 'female': 1}, inplace=True)
#replace missing values with the mode (most frequent value)
mode_cabin= df['Cabin'].mode()[0]
df['Cabin'].fillna(mode_cabin, inplace=True) 
df['Age'].fillna(df['Age'].median(), inplace=True)       
```

```python
X = df[['Pclass', 'Sex', 'Age']]
Y = df['Survived']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Initialize SVM
svm = SVC()
```

```python
# Perform Cross-Validation
cv_scores_svm = cross_val_score(svm, X, Y, cv=5)
print('SVM - Cross-Validation Scores: {cv_scores_svm}')
print('SVM - Mean CV Score: {cv_scores_svm.mean():}')
print('SVM - Standard Deviation of CV Scores: {cv_scores_svm.std():}')
```

Selecting the Best Model

```python

# Determine the best model
scores = {
    'Logistic Regression': cv_scores_log_reg.mean(),
    'Decision Tree': cv_scores_decision_tree.mean(),
    'Random Forest': cv_scores_random_forest.mean(),
    'SVM': cv_scores_svm.mean()
}

best_model = max(scores, key=scores.get)
print('Best Model: {best_model} with a Mean CV Score of {scores[best_model]:.2f}')
```
